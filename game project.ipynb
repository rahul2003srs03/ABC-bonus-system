```python
oct_start = pd.to_datetime('2022-10-01')
oct_end = pd.to_datetime('2022-10-31')
#by observing date
```

```python
# Total Number of Games Played per Player
games_played = df_gameplay.groupby("User ID")["Games Played"].count().reset_index()
games_played.columns = ["User ID", "Games Played"]
```

```python
# Total Deposit Amount and Count
deposit_agg = df_deposit.groupby("User ID")["Amount"].agg(["sum", "count"]).reset_index()
deposit_agg.columns = ["User ID", "total_deposit", "deposit_count"]
```

```python
# Total Withdrawal Amount and Count
withdraw_agg = df_withdrawal.groupby("User ID")["Amount"].agg(["sum", "count"]).reset_index()
withdraw_agg.columns = ["User ID", "total_withdraw", "withdraw_count"]
```

```python
loyalty_df = games_played.merge(deposit_agg, on="User ID", how="outer")
loyalty_df = loyalty_df.merge(withdraw_agg, on="User ID", how="outer")
```

```python
#From the above observation it was clear that user Id 0 neither deposite nor withdraw the amount 
# we fill 0 in place of NaN for calculation
loyalty_df.fillna(0, inplace=True)
print(loyalty_df.head())
```

```python
loyalty_df["loyalty_points"] = (
    (0.01 * loyalty_df["total_deposit"]) +
    (0.005 * loyalty_df["total_withdraw"]) +
    (0.001 * np.maximum((loyalty_df["deposit_count"] - loyalty_df["withdraw_count"]),0)) +
    (0.2 * loyalty_df["Games Played"]))
```

```python

# Correct way to parse your format
df_gameplay['Datetime'] = pd.to_datetime(df_gameplay['Datetime'], dayfirst=True, errors='coerce')
df_deposit['Datetime'] = pd.to_datetime(df_deposit['Datetime'], dayfirst=True, errors='coerce')
df_withdrawal['Datetime'] = pd.to_datetime(df_withdrawal['Datetime'], dayfirst=True, errors='coerce')


# Helper function to get slot mask
def get_slot_mask(df, date_str, slot):
    date = pd.to_datetime(date_str)
    if slot == 'S1':
        return (df['Datetime'] >= date) & (df['Datetime'] < date + pd.Timedelta(hours=12))
    elif slot == 'S2':
        return (df['Datetime'] >= date + pd.Timedelta(hours=12)) & (df['Datetime'] < date + pd.Timedelta(days=1))
```

```python
def calculate_slot_loyalty(date_str, slot):
    gp = df_gameplay[get_slot_mask(df_gameplay, date_str, slot)]
    dp = df_deposit[get_slot_mask(df_deposit, date_str, slot)]
    wd = df_withdrawal[get_slot_mask(df_withdrawal, date_str, slot)]

    gp_count = gp.groupby("User ID")["Games Played"].count().reset_index(name="Games Played")
    dp_agg = dp.groupby("User ID")["Amount"].agg(["sum", "count"]).reset_index()
    dp_agg.columns = ["User ID", "total_deposit", "deposit_count"]
    wd_agg = wd.groupby("User ID")["Amount"].agg(["sum", "count"]).reset_index()
    wd_agg.columns = ["User ID", "total_withdraw", "withdraw_count"]

    df = gp_count.merge(dp_agg, on="User ID", how="outer").merge(wd_agg, on="User ID", how="outer")
    df.fillna(0, inplace=True)

    df["loyalty_points"] = (
        0.01 * df["total_deposit"] +
        0.005 * df["total_withdraw"] +
        0.001 * (df["deposit_count"] - df["withdraw_count"]).clip(lower=0) +
        0.2 * df["Games Played"]
    )
    df["slot"] = f"{date_str} {slot}"
    return df[["User ID", "loyalty_points", "slot"]]
```

```python
slot_loyalties = pd.concat([
    calculate_slot_loyalty("2022-10-16", "S2"),
    calculate_slot_loyalty("2022-10-18", "S1"),
    calculate_slot_loyalty("2022-10-26", "S2"),
    calculate_slot_loyalty("2022-10-02", "S1"),
], ignore_index=True)
print(slot_loyalties)
```

```python
gp_oct = df_gameplay[(df_gameplay["Datetime"] >= oct_start) & (df_gameplay["Datetime"] < oct_end)]
dp_oct = df_deposit[(df_deposit["Datetime"] >= oct_start) & (df_deposit["Datetime"] < oct_end)]
wd_oct = df_withdrawal[(df_withdrawal["Datetime"] >= oct_start) & (df_withdrawal["Datetime"] < oct_end)]
```

```python
gp_m = gp_oct.groupby("User ID")["Games Played"].count().reset_index(name="Games Played")
dp_m = dp_oct.groupby("User ID")["Amount"].agg(["sum", "count"]).reset_index()
dp_m.columns = ["User ID", "total_deposit", "deposit_count"]
wd_m = wd_oct.groupby("User ID")["Amount"].agg(["sum", "count"]).reset_index()
wd_m.columns = ["User ID", "total_withdraw", "withdraw_count"]
monthly_df = gp_m.merge(dp_m, on="User ID", how="outer").merge(wd_m, on="User ID", how="outer")
monthly_df.fillna(0, inplace=True)
```

```python
monthly_df["loyalty_points"] = (
    0.01 * monthly_df["total_deposit"] +
    0.005 * monthly_df["total_withdraw"] +
    0.001 * (monthly_df["deposit_count"] - monthly_df["withdraw_count"]).clip(lower=0) +
    0.2 * monthly_df["Games Played"]
)
monthly_df["rank"] = monthly_df.sort_values(
    by=["loyalty_points", "Games Played"], 
    ascending=[False, False]
).reset_index(drop=True).index + 1
```

```python
avg_deposit = df_deposit["Amount"].mean()
avg_deposit_per_user_oct = dp_oct.groupby("User ID")["Amount"].sum().mean()
avg_games_per_user_oct = gp_oct.groupby("User ID")["Games Played"].count().mean()
```

```python
results_summary = {
    "slot_loyalties": slot_loyalties,
    "monthly_loyalty_ranking": monthly_df.sort_values(by="rank").reset_index(drop=True),
    "average_deposit": avg_deposit,
    "avg_deposit_per_user_oct": avg_deposit_per_user_oct,
    "avg_games_per_user_oct": avg_games_per_user_oct
}

results_summary
```

It is heavy bias toward deposite amount means less activity but high deposite make user on top
No reward for game wins or duration like: if a player play 20 game and losing all games has same reward as the winner get no difference

**suggested formula**
Loyalty Points =
